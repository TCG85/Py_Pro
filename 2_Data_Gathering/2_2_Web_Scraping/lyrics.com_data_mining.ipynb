##  important Libraries

```python
import requests
import re
import time
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

page = requests.get('https://www.lyrics.com/artist/Marilyn-Manson/45021', headers=headers)
html = page.text
```

```python
html
```

```python
pattern = r'href=\"(/lyric[/+\w]+)'
```

```python
sub_link = re.findall(pattern, html)
sub_link
```

```python
artist = 'Marilyn_Manson'
n = 0
for i in sub_link[:10]:
    base = 'https://www.lyrics.com'
    url = base + i
    result = requests.get(url)
    name = url.split('/')[-1]
    name = name. replace('+','')
    f = open(f'lyrics_{artist}/{n}-{artist}-{name}.html', 'w', encoding='utf-8')
    f.write(result.text)
    f.close()
    print(f'{n}-{artist}-{name}.html created')
    n += 1
    time.sleep(10)
    
```
